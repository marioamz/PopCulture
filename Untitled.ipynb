```python
import pandas as pd
import re
import wikipedia as wiki
import jellyfish as j
from bs4 import BeautifulSoup
```

```python
books_df = pd.read_excel("NYTimesBestSellers.xlsx")
```

```python
books_df[:20]
```

```python
authors = books_df["Author"].unique()
```

```python
def find_best_match(author_name):
    actual_name = None
    most_sim = 0
    wsearch = wiki.search(author_name)
    for result in wsearch:
        jwscore = j.jaro_winkler(author_name, result)
        if jwscore > 0.9 and jwscore > most_sim:
            actual_name = result
            most_sim = jwscore
    return actual_name
```

```python
for author in authors:
    actual_name = find_best_match(author)
    to_clean = []
    if author != actual_name: 
        if actual_name is not None:
            print(author, actual_name)
```
