```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
from plotly.graph_objs import *
import math

pd.options.mode.chained_assignment = None
pd.options.display.max_columns = 999
```

```python
FEELS = ["anticipation", "anger", "joy", "trust", "fear", "surprise", "sadness", "disgust"]


def scatter_trace(emotion, df):
    
    # filter by emotion
    emo = "{} percentage".format(emotion)
    df = df[['Year', emo]].sort_values(by='Year')
    lab = emotion.capitalize()
    trace = go.Scatter(x=df['Year'],
                       y=df[emo],
                       name=lab,
                       mode='lines',
                       line=dict(shape='spline'))
    return trace


def scatter_plot(df):
    data = [scatter_trace(f, df) for f in FEELS]
    layout = Layout(title="Evolution of Emotions",
                    xaxis=dict(title='Year',
                               rangeslider=dict()),
                    yaxis=dict(title='%'))
    fig = Figure(data=data, layout=layout)

    py.iplot(fig)
```

```python
def plot_data(filename):
    df = pd.read_csv(filename, header='infer')
    scatter_plot(df)

    
    
    
```
